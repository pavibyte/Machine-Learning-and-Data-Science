```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

HOUSE PRICE DISTRIBUTION

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'],bins=30,kde=True)
plt.title("Distribution of house prices")
plt.show()
```

OBSERVE:

Is the distribution symmetric or skewed?

The distribution is skewed. It is right skewed.

Which side is the tail on?

The tail is on the right side.

BOXPLOT FOR OUTLIERS

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=df['price'])
plt.title("boxplot of house prices")
plt.show()
```

AREA VS PRICE(intuition test)

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x=df['area'],y=df['price'])
plt.title("Living area vs Sale price")
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

Does price generally increase with area?

No, price doesn't generally increase with area.

Are there exceptions?

Yes, for some houses where area is less the price is less too. But there are cases where even for smaller area the prices are higher.

Any weird points?

the wierd points are some with less area and more price and the vise versa datapoints.
